```python
import json
catagory = "50%"
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'SimSun'  # SimSun 是一個常用的支持中文字符的字體
```

```python
with open(f'log\commom_word-25%_indicator.json','r') as f:
    indicater = json.load(f) 
```

```python
point = []
index = []
for tag, value in indicater.items():
    index.append((int(tag.split('-')[0])*6400 + int(tag.split('-')[1]))//100)
    point.append((value['f1']))
```

```python
plt.plot(index, point, label='')

# 加入標題和標籤
plt.title('')
plt.xlabel('')
plt.ylabel('')

# 顯示圖例
plt.legend()

# 顯示圖表
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 資料
categories =  ['ㄅ','ㄆ','ㄇ','ㄈ','ㄉ','ㄊ','ㄋ','ㄌ','ㄍ','ㄎ','ㄏ','ㄐ',
               'ㄑ','ㄒ','ㄓ','ㄔ','ㄕ','ㄖ','ㄗ','ㄘ','ㄙ','ㄚ','ㄛ','ㄜ',
               'ㄝ','ㄞ','ㄟ','ㄠ','ㄡ','ㄢ','ㄣ','ㄤ','ㄥ','ㄦ','ㄧ','ㄨ','ㄩ']
values_75 = [10,8,10,9,10,10,7,10,10,10,10,10,10,10,10,8,10,5,10,10,5,10,5,8,10,10,10,10,9,1,10,6,10,10,4,10,10]
values_50 = [10,8,10,9,10,10,6,10,10,10,10,10,10,10,10,10,10,6,10,10,7,10,5,10,10,10,10,10,7,1,10,5,10,10,5,10,10]
values_25 = [10,7,10,8,9, 10,5,10,8,10,10,10,10,10,10,8,10,7,10,10,5,10,10,6,10,10,10,7,7,3,8,8,10,10,4,10,10]

values1 = np.array(values_75)
values2 = np.array(values_50)
values3 = np.array(values_25)

# 調整寬度
bar_width = 0.15  # 設定每個長條的寬度
plt.figure(figsize=(16,5))
# 設定位置
bar_positions1 = np.arange(len(categories))
bar_positions2 = bar_positions1 + bar_width
bar_positions3 = bar_positions2 + bar_width

# 畫長條圖
plt.bar(bar_positions1, values1, width=bar_width, label='75%')
plt.bar(bar_positions2, values2, width=bar_width, label='50%')
plt.bar(bar_positions3, values3, width=bar_width, label='25%')

# 加上標題和軸標籤
plt.title('25%、50%、75%')
plt.xlabel('類別')
plt.ylabel('辨識成功次數')
# 設定 x 軸標籤
plt.xticks(bar_positions2, categories)

# 加上圖例
plt.legend()

# 顯示圖表
plt.show()
```
