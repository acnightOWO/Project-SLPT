# rename


```python
def keep_numbers_and_minus(input_string):
    return ''.join([char for char in input_string if char.isdigit() or char == '-'])
```

```python
Database_dir = '/content/Dataset'
file_list = os.listdir(Database_dir)

for file in file_list:
    try:
        new_name = keep_numbers_and_minus(os.path.splitext(os.path.basename(file))[0].replace('-',' ').replace(' ','-',3))
        new_path = os.path.join(Database_dir, new_name) + os.path.splitext(file)[-1]
        os.rename(os.path.join(Database_dir, file), new_path)
    except:
        pass
```
