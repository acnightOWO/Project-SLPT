```python
import cv2
import mediapipe as mp
import numpy as np

# 初始化MediaPipe Hand模块
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(min_tracking_confidence=0.4,
                       min_detection_confidence=0.1,
                       max_num_hands=2,
                       static_image_mode=False)

# 打开视频文件的函数
def display(video_path):
    cap = cv2.VideoCapture(video_path)
    i2=0
    while cap.isOpened():
        ret, oframe = cap.read()
        if not ret:
            break

        # 将帧水平翻转以便后续显示
        frame = cv2.flip(oframe, 1)
        frame = frame[300:880, 40:520]
        white_image = np.full((580, 480, 3), 255, dtype=np.uint8)

        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # 处理帧并获取手部关键点
        results = hands.process(rgb_frame)

        # 检查每只手的类型（左手或右手）
        hand_types = []
        if results.multi_handedness:
            for handedness in results.multi_handedness:
                hand_type = handedness.classification[0].label
                hand_types.append(hand_type)

        # 检查是否检测到关键点
        if results.multi_hand_landmarks:
            for i, landmarks in enumerate(results.multi_hand_landmarks):
                # 根据手的类型选择颜色
                color = (0, 0, 255) if hand_types[i] == 'Right' else (255, 0, 0)

                # 绘制关节点连接线
                hand_connections = [
                    (0, 1), (1, 2), (2, 3), (3, 4),  # Thumb
                    (0, 5), (5, 6), (6, 7), (7, 8),  # Index finger
                    (0, 9), (9, 10), (10, 11), (11, 12),  # Middle finger
                    (0, 13), (13, 14), (14, 15), (15, 16),  # Ring finger
                    (0, 17), (17, 18), (18, 19), (19, 20)  # Little finger
                ]

                for connection in hand_connections:
                    point1 = landmarks.landmark[connection[0]]
                    point2 = landmarks.landmark[connection[1]]
                    x1, y1 = int(point1.x * frame.shape[1]), int(point1.y * frame.shape[0])
                    x2, y2 = int(point2.x * frame.shape[1]), int(point2.y * frame.shape[0])
                    cv2.line(frame, (x1, y1), (x2, y2), (0,255,0), 2)
                    cv2.line(white_image, (x1, y1), (x2, y2), (0,255,0), 2)

                # 绘制关节点
                for landmark in landmarks.landmark:
                    x, y = int(landmark.x * frame.shape[1]), int(landmark.y * frame.shape[0])
                    cv2.circle(frame, (x, y), 2, color, -1)
                    cv2.circle(white_image, (x, y), 2, color, -1)

        # 显示结果帧
        cv2.imshow("Hand Landmarks", frame)
        cv2.imshow("Hand ", white_image)
        cv2.imwrite(fr'C:\Users\Huang\Pictures\paper picture\video\white-{i2}.jpg', white_image)
        cv2.imwrite(fr'C:\Users\Huang\Pictures\paper picture\video\Hand Landmarks-{i2}.jpg', frame)
        cv2.imwrite(fr'C:\Users\Huang\Pictures\paper picture\video\back-{i2}.jpg', oframe)
    # 释放资源
        i2+=1
        # 当按下'q'键时退出循环
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

# 调用函数并传入视频文件路径
display(r"C:\Users\Huang\Pictures\paper picture\video\output_video.mp4")
```
