```python
import json
import os

with open('word.json', 'r') as f:
    words_json = json.load(f)
with open('signs_videos_chosen.json', 'r') as q:
    sentences_json = json.load(q)
```

```python
words = []
for i,word in enumerate(words_json):
    processed_word = word['elemtext']
    processed_word = processed_word.split('\n')
    if processed_word == word['elemtext']:
        words.append((processed_word,i))
    else:
        for o,oword in enumerate(processed_word):
            if oword.isdigit():
                pass
            else:
                words.append((oword,i))
print(words)
```

```python
output = []
for i,sentence in enumerate(sentences_json):#sentence是句子
    processed_sentence = sentence['elemsplit'].split('/')
    index = []
    for o,osentence in enumerate(processed_sentence):#osectence是單詞
        wordindex = 'none'
        for o,oword in enumerate(words):
            if oword[0] == osentence:
                wordindex = words_json[oword[1]]['url']
        index.append(wordindex)
    output.append(index)
print(output)
with open('output.json','w') as file:
    json.dump(output,file,indent=4)
```

```python
finale_output = []
for i,sentence in enumerate(sentences_json):#sentence是句子
    sentences_json[i]['sep_videos'] = output[i]
with open('finale_output.json','w') as file:
    json.dump(sentences_json,file,indent=4,ensure_ascii=False)
```
